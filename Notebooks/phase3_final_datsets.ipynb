### Add Possession-Relative Features

Transform features to be relative to the possessing team (team_id), not fixed home/away perspective.

```python
# Add home team identification
df_events['home_team_id'] = df_events['game_id'].map(home_team_lookup)

# Binary flag: Is the possessing team (team_id) playing at home?
df_events['home_team_poss'] = (df_events['team_id'] == df_events['home_team_id']).astype(int)

print(f"Added home_team_poss flag:")
print(f"  Home possession: {df_events['home_team_poss'].sum():,} events ({df_events['home_team_poss'].mean()*100:.1f}%)")
print(f"  Away possession: {(1-df_events['home_team_poss']).sum():,} events ({(1-df_events['home_team_poss']).mean()*100:.1f}%)")

# Calculate period_time_remaining (20 min periods = 1200 seconds)
df_events['period_time_remaining'] = 1200 - df_events['period_time']

#clip to 0-1200 range to avoid any potential outliers (e.g. due to tracking errors)
df_events['period_time_remaining'] = df_events['period_time_remaining'].clip(0, 1200)

print(f"\n✓ Calculated period_time_remaining")
print(f"  Range: [{df_events['period_time_remaining'].min():.1f}, {df_events['period_time_remaining'].max():.1f}]")

# Transform score differential to possession-relative
# If team_id is home: use score_differential_home
# If team_id is away: use score_differential_away (which is negative of home)
df_events['score_differential_poss'] = np.where(
    df_events['home_team_poss'] == 1,
    df_events['score_differential_home'],
    df_events['score_differential_away']
)

print(f"\n✓ Calculated score_differential_poss (relative to possessing team)")
print(f"  Range: [{df_events['score_differential_poss'].min()}, {df_events['score_differential_poss'].max()}]")
print(f"  Mean: {df_events['score_differential_poss'].mean():.3f}")

# Transform skater counts to possession-relative
df_events['n_skaters_poss'] = np.where(
    df_events['home_team_poss'] == 1,
    df_events['n_home_skaters'],
    df_events['n_away_skaters']
)

df_events['n_skaters_def'] = np.where(
    df_events['home_team_poss'] == 1,
    df_events['n_away_skaters'],
    df_events['n_home_skaters']
)

print(f"\n✓ Calculated n_skaters_poss and n_skaters_def")
print(f"  Possession skaters - Mean: {df_events['n_skaters_poss'].mean():.2f}, Range: [{df_events['n_skaters_poss'].min()}, {df_events['n_skaters_poss'].max()}]")
print(f"  Defending skaters - Mean: {df_events['n_skaters_def'].mean():.2f}, Range: [{df_events['n_skaters_def'].min()}, {df_events['n_skaters_def'].max()}]")

# Transform net empty flags to possession-relative
df_events['net_empty_poss'] = np.where(
    df_events['home_team_poss'] == 1,
    df_events['is_home_net_empty'],
    df_events['is_away_net_empty']
).astype(int)

df_events['net_empty_def'] = np.where(
    df_events['home_team_poss'] == 1,
    df_events['is_away_net_empty'],
    df_events['is_home_net_empty']
).astype(int)

print(f"\n✓ Calculated net_empty_poss and net_empty_def")
print(f"  Possessing net empty: {df_events['net_empty_poss'].sum():,} events ({df_events['net_empty_poss'].mean()*100:.2f}%)")
print(f"  Defending net empty: {df_events['net_empty_def'].sum():,} events ({df_events['net_empty_def'].mean()*100:.2f}%)")

print(f"\n{'='*80}")
print(f"POSSESSION-RELATIVE TRANSFORMATIONS COMPLETE")
print(f"{'='*80}")
print(f"\nSample of transformed features:")
sample_cols = ['team_id', 'home_team_id', 'home_team_poss', 
               'score_differential_poss', 'n_skaters_poss', 'n_skaters_def',
               'net_empty_poss', 'net_empty_def', 'period_time_remaining']
print(df_events[sample_cols].head(10))
```

### Define Feature Groups for Model

Based on the 4-branch architecture in README:
- **Categorical Features**: event_type, outcome, period (will be integer-embedded separately in model)
- **Text Features**: embedding_text_clean (cleaned, leakage-checked text to be pre-embedded to 384-dim)
- **Continuous Event Features**: Raw coordinates/angles + Possession-relative features (score_differential_poss, n_skaters_poss/def, period_time_remaining)
- **Binary Features**: home_team_poss, net_empty_poss, net_empty_def, is_eos (EOS token for sequence terminators)
- **Tracking Features**: 72 dimensions (actor:6 + teammates:30 + opponents:36)

Note: Features will be transformed to possession-relative perspective (relative to team_id, not fixed home/away).

```python
# Define feature groups for final dataset
# These will be created through transformations below

# Categorical features (will be integer-embedded separately)
CATEGORICAL_FEATURES = ['event_type', 'outcome', 'period']

# Text features (will be pre-embedded with SentenceTransformer)
TEXT_FEATURES = ['embedding_text_clean']

# Continuous event features (possession-relative, to be calculated)
CONTINUOUS_EVENT_FEATURES_RAW = [
    'x',
    'y',
    'x_adj', 
    'y_adj', 
    'dest_x_adj',
    'dest_y_adj',
    'period_time',
    'distance_to_net_event', 
    'angle_to_net_event'
]

CONTINUOUS_POSSESSION_FEATURES = [
    'period_time_remaining',  # 1200 - period_time
    'score_differential_poss',  # Relative to possessing team
    'n_skaters_poss',  # Number of skaters for possessing team
    'n_skaters_def'  # Number of skaters for defending team
]

CONTINUOUS_EVENT_FEATURES = CONTINUOUS_EVENT_FEATURES_RAW + CONTINUOUS_POSSESSION_FEATURES

# Binary features (possession-relative, to be calculated)
BINARY_EVENT_FEATURES = [
    'net_empty_poss',  # Is possessing team's net empty
    'net_empty_def',  # Is defending team's net empty
    'home_team_poss'  # Is possessing team playing at home
]

# Tracking features (72 dimensions) - already exist
# Actor: 6 features
ACTOR_FEATURES = [
    'actor_rel_x', 'actor_rel_y', 'actor_vx_std', 'actor_vy_std', 
    'actor_mask', 'actor_is_imputed'
]

# Teammates: 5 players × 6 features = 30
TEAMMATE_FEATURES = []
for i in range(1, 6):
    TEAMMATE_FEATURES.extend([
        f'tm_{i}_rel_x', f'tm_{i}_rel_y', f'tm_{i}_vx_std', f'tm_{i}_vy_std',
        f'tm_{i}_distance', f'tm_{i}_mask'
    ])

# Opponents: 6 players × 6 features = 36
OPPONENT_FEATURES = []
for i in range(1, 7):
    OPPONENT_FEATURES.extend([
        f'opp_{i}_rel_x', f'opp_{i}_rel_y', f'opp_{i}_vx_std', f'opp_{i}_vy_std',
        f'opp_{i}_distance', f'opp_{i}_mask'
    ])

# All tracking features
TRACKING_FEATURES = ACTOR_FEATURES + TEAMMATE_FEATURES + OPPONENT_FEATURES

print(f"Feature group sizes (after transformations will be applied):")
print(f"  Categorical: {len(CATEGORICAL_FEATURES)}")
print(f"  Text: {len(TEXT_FEATURES)} (will be SentenceTransformer embedded)")
print(f"  Continuous Event: {len(CONTINUOUS_EVENT_FEATURES)}")
print(f"    - Raw coordinates/angles: {len(CONTINUOUS_EVENT_FEATURES_RAW)}")
print(f"    - Possession-relative: {len(CONTINUOUS_POSSESSION_FEATURES)}")
print(f"  Binary: {len(BINARY_EVENT_FEATURES)}")
print(f"  Tracking Features: {len(TRACKING_FEATURES)}")
print(f"    - Actor: {len(ACTOR_FEATURES)}")
print(f"    - Teammates: {len(TEAMMATE_FEATURES)}")
print(f"    - Opponents: {len(OPPONENT_FEATURES)}")

# Verify raw columns exist now
print(f"\nVerifying currently available columns...")
required_now = CATEGORICAL_FEATURES + CONTINUOUS_EVENT_FEATURES_RAW + TEXT_FEATURES + TRACKING_FEATURES
missing_cols = [col for col in required_now if col not in df_events.columns]

if missing_cols:
    print(f"❌ Missing columns: {missing_cols}")
else:
    print(f"✓ All required raw columns present")
    print(f"  (Possession-relative features will be calculated next)")
```

```python
import re

# Install sentence-transformers if not available
try:
    from sentence_transformers import SentenceTransformer
    print("✓ sentence-transformers already installed")
except ImportError:
    print("Installing sentence-transformers...")
    import subprocess
    subprocess.check_call(['pip', 'install', 'sentence-transformers'])
    from sentence_transformers import SentenceTransformer
    print("✓ sentence-transformers installed")

import torch
try:
    import intel_extension_for_pytorch as ipex
    print(f"✓ Intel Extension for PyTorch available (version: {ipex.__version__})")
except ImportError:
    print("⚠ Intel Extension not found (install 'intel_extension_for_pytorch' for Arc GPU acceleration)")

if 'embedding_text_clean' not in df_events.columns:
    raise KeyError("Missing required column: embedding_text_clean")

# Double-check for leakage before embedding
leak_regex = re.compile(
    r"\bgoal\b|\bgoals\b|\bon\s*net\b|\bonnet\b|\bwith\s*goal\b|\bwith\s*shot\s*on\s*net(?:\s*whistle)?\bwith*\b",
    flags=re.IGNORECASE
)
blocked_compact_tokens = {'withgoal', 'withshotonnet', 'withshotonnetwhistle'}

embedding_text_series = df_events['embedding_text_clean'].fillna('').astype(str).str.strip()
compact_series = embedding_text_series.str.lower().str.replace(r'[^a-z0-9]', '', regex=True)
leak_mask = embedding_text_series.str.contains(leak_regex, na=False) | compact_series.isin(blocked_compact_tokens)

n_leaks = int(leak_mask.sum())
print(f"Leakage check before embedding: {n_leaks:,} rows flagged")
if n_leaks > 0:
    print("Sample leaked rows (up to 10):")
    print(df_events.loc[leak_mask, ['game_id', 'sequence_id', 'sl_event_id', 'embedding_text_clean']].head(10).to_string(index=False))
    raise ValueError("Leakage tokens detected in embedding_text_clean. Clean these rows before embedding.")

# Select device for text embedding
if torch.xpu.is_available():
    device = torch.device("xpu")
    print("✓ Using Intel XPU (Arc GPU) for text embeddings")
elif torch.cuda.is_available():
    device = torch.device("cuda")
    print("✓ Using CUDA GPU for text embeddings")
else:
    device = torch.device("cpu")
    print("⚠ Using CPU for text embeddings (will be slower)")

# Load sentence transformer model
print(f"\nLoading SentenceTransformer model: all-MiniLM-L6-v2...")
model = SentenceTransformer('all-MiniLM-L6-v2', device=str(device))
print(f"✓ Model loaded (embedding dimension: {model.get_sentence_embedding_dimension()})")

# Use embedding_text_clean directly for text embeddings
df_events['embedding_text_clean'] = embedding_text_series
unique_texts = pd.unique(df_events['embedding_text_clean'])

print(f"\n✓ Prepared embedding_text_clean for encoding")
print(f"  Total events: {len(df_events):,}")
print(f"  Unique text combinations: {len(unique_texts):,}")

# Encode unique texts (with normalized embeddings)
print(f"\nEncoding unique texts (this may take a few minutes)...")
print(f"  Device: {device}")
print(f"  Batch size: 32")

unique_embeddings = model.encode(
    unique_texts.tolist(),
    batch_size=32,
    show_progress_bar=True,
    normalize_embeddings=True,
    convert_to_numpy=True
)

print(f"\n✓ Encoded {len(unique_texts):,} unique texts")
print(f"  Embedding shape: {unique_embeddings.shape}")

# Create mapping from text to embedding index
text_to_idx = {text: idx for idx, text in enumerate(unique_texts)}

# Map all events to embedding indices (for final dataset join-back)
df_events['text_embedding_idx'] = df_events['embedding_text_clean'].map(text_to_idx).astype('int32')

print(f"\n✓ Mapped all events to embedding indices")
print(f"  Index range: [0, {df_events['text_embedding_idx'].max()}]")

# Save embeddings as separate .npy file
embeddings_output_path = OUTPUT_DIR / "text_embeddings.npy"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
np.save(embeddings_output_path, unique_embeddings.astype(np.float32))

print(f"\n✓ Saved text embeddings to: {embeddings_output_path}")
print(f"  Shape: {unique_embeddings.shape}")
print(f"  Size: {embeddings_output_path.stat().st_size / 1024 / 1024:.2f} MB")
print(f"  Format: float32, L2-normalized")

# Save explicit embedding index map for traceability
embedding_map_path = OUTPUT_DIR / "text_embedding_idx_map.parquet"
pd.DataFrame({
    'text_embedding_idx': np.arange(len(unique_texts), dtype=np.int32),
    'embedding_text_clean': unique_texts
}).to_parquet(embedding_map_path, index=False)
print(f"✓ Saved embedding index map to: {embedding_map_path}")

# Save updated events dataframe with embedding index
events_output_path = OUTPUT_DIR / "events_with_embedding_indices.parquet"
df_events.to_parquet(events_output_path, index=False)
print(f"\n✓ Saved updated events dataframe with embedding indices to: {events_output_path}")
print(f"  Shape: {df_events.shape}")
print(f"  Size: {events_output_path.stat().st_size / 1024 / 1024:.2f} MB")

# Verify sample embedding
sample_idx = 0
sample_text = unique_texts[sample_idx]
sample_embedding = unique_embeddings[sample_idx]
print(f"\nSample embedding:")
print(f"  Text: '{sample_text[:100]}...'")
print(f"  Embedding norm: {np.linalg.norm(sample_embedding):.6f} (should be ≈1.0)")
print(f"  First 5 dims: {sample_embedding[:5]}")
```

### Add EOS (End-of-Sequence) Tokens

Add explicit EOS tokens for sequence terminators (whistle, goal, period end, etc.) to help model learn sequence boundaries.

```python
# Define sequence terminator event types that should get EOS token
TERMINATOR_EVENTS = [
    'whistle',
    'goal',
    'end_of_period',  # Note: matches actual event_type name in data
]

# Add is_eos flag for last event in each sequence
df_events['is_last_in_seq'] = False
last_event_idx = df_events.groupby(['game_id', 'sequence_id']).tail(1).index
df_events.loc[last_event_idx, 'is_last_in_seq'] = True

# Mark as EOS if it's last event AND is a terminator type
df_events['is_eos'] = (
    df_events['is_last_in_seq'] & 
    df_events['event_type'].isin(TERMINATOR_EVENTS)
).astype(int)

print(f"EOS Token Statistics:")
print(f"{'='*80}")
print(f"  Total events: {len(df_events):,}")
print(f"  Last events in sequences: {df_events['is_last_in_seq'].sum():,}")
print(f"  EOS tokens (terminator as last): {df_events['is_eos'].sum():,}")
print(f"  EOS percentage: {df_events['is_eos'].mean()*100:.2f}%")

print(f"\nEOS tokens by event type:")
eos_by_type = df_events[df_events['is_eos'] == 1]['event_type'].value_counts()
for event_type, count in eos_by_type.items():
    pct = count / df_events['is_eos'].sum() * 100
    print(f"  {event_type:20s}: {count:>8,} ({pct:>5.1f}%)")

print(f"\n✓ Added is_eos flag (will be used as binary feature in model)")
```

```python
# Dynamic-horizon target engineering (continuous-state framing)
# ------------------------------------------------------------------
# Target classes (actor-relative):
#   0 = possession team scores within horizon
#   1 = opposing team scores within horizon
#   2 = no goal within horizon

print("=" * 80)
print("DYNAMIC HORIZON TARGET ENGINEERING (ACTOR-RELATIVE, 3-CLASS)")
print("=" * 80)

CLASS_POSS_GOAL = 0
CLASS_DEF_GOAL = 1
CLASS_NO_GOAL = 2

# Hard schema gate at target step as well (defensive against partial reruns)
required_target_cols = [
    'game_id', 'period', 'period_time', 'event_type', 'team_id',
    'player_id', 'detail', 'sequence_id', 'sl_event_id',
    'game_time_sec', 'is_boundary_event', 'is_end_of_period'
 ]
missing_target_cols = [col for col in required_target_cols if col not in df_events.columns]
if missing_target_cols:
    raise KeyError(
        "Phase 3 hard-fail: missing required columns for dynamic target engineering: "
        f"{missing_target_cols}. Re-run updated Phase 2/earlier Phase 3 cells."
    )

# Preserve current ordering unless abnormal temporal drift is detected
# Abnormal = period_time delta < -1 within (game_id, period, sequence_id)
df_events['period_time'] = pd.to_numeric(df_events['period_time'], errors='coerce')
seq_keys = ['game_id', 'period', 'sequence_id']
delta = df_events.groupby(seq_keys, sort=False)['period_time'].diff()
abnormal_mask = delta < -1
n_abnormal = int(abnormal_mask.sum())
n_abnormal_seq = int(df_events.loc[abnormal_mask, seq_keys].drop_duplicates().shape[0])
print(f"Abnormal period_time deltas (< -1): {n_abnormal:,} across {n_abnormal_seq:,} sequences")

if n_abnormal > 0:
    time_sort_cols = ['game_id', 'period', 'game_time_sec', 'period_time']
    if 'sl_event_id' in df_events.columns:
        time_sort_cols.append('sl_event_id')
    if 'sequence_event_id' in df_events.columns:
        time_sort_cols.append('sequence_event_id')
    df_events = df_events.sort_values(time_sort_cols, kind='mergesort').reset_index(drop=True)
    print("Applied corrective temporal resort for dynamic horizon due to abnormality.")
else:
    df_events = df_events.reset_index(drop=True)
    print("No abnormality detected; preserving current event ordering for dynamic horizon.")
df_events['_row_id'] = np.arange(len(df_events), dtype=np.int64)

# Base rows eligible for temporal lookups (period_time required)
base_times = df_events[['_row_id', 'game_id', 'period', 'period_time']].copy()
base_times = base_times.dropna(subset=['period_time'])

def assign_next_time_by_group(base_df, ref_df, value_col):
    """Assign strictly next event time per (game_id, period) using searchsorted."""
    out = np.full(len(df_events), np.nan, dtype=float)
    if base_df.empty or ref_df.empty:
        return out

    ref_groups = {
        key: grp[value_col].to_numpy(dtype=float)
        for key, grp in ref_df.groupby(['game_id', 'period'], sort=False)
    }

    for key, grp in base_df.groupby(['game_id', 'period'], sort=False):
        ref_times = ref_groups.get(key)
        if ref_times is None or len(ref_times) == 0:
            continue

        t = grp['period_time'].to_numpy(dtype=float)
        idx = np.searchsorted(ref_times, t, side='right')  # strictly future
        valid = idx < len(ref_times)
        if not np.any(valid):
            continue

        row_ids = grp['_row_id'].to_numpy(dtype=np.int64)[valid]
        out[row_ids] = ref_times[idx[valid]]

    return out

def assign_next_goal_by_group(base_df, goals_df):
    """Assign strictly next goal time and team per (game_id, period) using searchsorted."""
    out_time = np.full(len(df_events), np.nan, dtype=float)
    out_team = np.empty(len(df_events), dtype=object)
    out_team[:] = np.nan
    if base_df.empty or goals_df.empty:
        return out_time, out_team

    grouped_goals = {
        key: (
            grp['next_goal_time'].to_numpy(dtype=float),
            grp['next_goal_team_id'].to_numpy(dtype=object)
        )
        for key, grp in goals_df.groupby(['game_id', 'period'], sort=False)
    }

    for key, grp in base_df.groupby(['game_id', 'period'], sort=False):
        goal_pack = grouped_goals.get(key)
        if goal_pack is None:
            continue

        goal_times, goal_teams = goal_pack
        if len(goal_times) == 0:
            continue

        t = grp['period_time'].to_numpy(dtype=float)
        idx = np.searchsorted(goal_times, t, side='right')  # strictly future
        valid = idx < len(goal_times)
        if not np.any(valid):
            continue

        row_ids = grp['_row_id'].to_numpy(dtype=np.int64)[valid]
        out_time[row_ids] = goal_times[idx[valid]]
        out_team[row_ids] = goal_teams[idx[valid]]

    return out_time, out_team

# ------------------------------------------------------------------
# 1) Build dynamic horizon_end_time
#    Rule A: default +20s
#    Rule B: penalty + penaltydrawn use next faceoff + 120s (no severity split)
#    Rule C: cap by period max time
# ------------------------------------------------------------------
PERIOD_MAX_FALLBACK = 1200.0
DEFAULT_HORIZON_SEC = 20.0
PENALTY_WINDOW_SEC = 120.0

# Per-(game, period) cap (no cross-period bleeding)
period_cap = df_events.groupby(['game_id', 'period'])['period_time'].transform('max')
period_cap = period_cap.clip(upper=PERIOD_MAX_FALLBACK)
df_events['period_cap_time'] = period_cap

# Default horizon
df_events['horizon_end_time'] = df_events['period_time'] + DEFAULT_HORIZON_SEC

# Next faceoff time (strictly future within same game/period)
faceoffs = df_events[df_events['event_type'] == 'faceoff'][['game_id', 'period', 'period_time']].copy()
faceoffs = faceoffs.dropna(subset=['period_time'])
faceoffs = faceoffs.rename(columns={'period_time': 'next_faceoff_time'})
faceoffs = faceoffs.sort_values(['game_id', 'period', 'next_faceoff_time'], kind='mergesort')

df_events['next_faceoff_time'] = assign_next_time_by_group(base_times, faceoffs, 'next_faceoff_time')

# penalty + penaltydrawn receive the same 120-second extension from next faceoff
is_penalty_window_event = df_events['event_type'].isin(['penalty', 'penaltydrawn'])
df_events['penalty_window_sec'] = np.where(is_penalty_window_event, PENALTY_WINDOW_SEC, np.nan)

has_faceoff_for_penalty = is_penalty_window_event & df_events['next_faceoff_time'].notna()
df_events.loc[has_faceoff_for_penalty, 'horizon_end_time'] = (
    df_events.loc[has_faceoff_for_penalty, 'next_faceoff_time']
    + df_events.loc[has_faceoff_for_penalty, 'penalty_window_sec']
)

# Cap horizon at period boundary
df_events['horizon_end_time'] = np.minimum(df_events['horizon_end_time'], df_events['period_cap_time'])

# ------------------------------------------------------------------
# 2) sec_next_stoppage (next whistle/goal/end_of_period)
# ------------------------------------------------------------------
stoppages = df_events[
    df_events['event_type'].isin(['whistle', 'goal', 'end_of_period'])
][['game_id', 'period', 'period_time']].copy()
stoppages = stoppages.dropna(subset=['period_time'])
stoppages = stoppages.rename(columns={'period_time': 'next_stoppage_time'})
stoppages = stoppages.sort_values(['game_id', 'period', 'next_stoppage_time'], kind='mergesort')

df_events['next_stoppage_time'] = assign_next_time_by_group(base_times, stoppages, 'next_stoppage_time')
df_events['sec_next_stoppage'] = df_events['next_stoppage_time'] - df_events['period_time']

# ------------------------------------------------------------------
# 3) Dynamic-window goal evaluation using earliest future goal priority
# ------------------------------------------------------------------
goals = df_events[df_events['event_type'] == 'goal'][['game_id', 'period', 'period_time', 'team_id']].copy()
goals = goals.dropna(subset=['period_time'])
goals = goals.rename(columns={'period_time': 'next_goal_time', 'team_id': 'next_goal_team_id'})
goals = goals.sort_values(['game_id', 'period', 'next_goal_time'], kind='mergesort')

next_goal_time, next_goal_team = assign_next_goal_by_group(base_times, goals)
df_events['next_goal_time'] = next_goal_time
df_events['next_goal_team_id'] = next_goal_team

in_dynamic_horizon = (
    df_events['next_goal_time'].notna()
    & (df_events['next_goal_time'] <= df_events['horizon_end_time'])
)

# Actor-relative 3-class target
if 'target' in df_events.columns:
    df_events = df_events.drop(columns=['target'])

df_events['target'] = CLASS_NO_GOAL

valid_actor_team = df_events['team_id'].notna()
poss_goal_mask = in_dynamic_horizon & valid_actor_team & (df_events['next_goal_team_id'] == df_events['team_id'])
def_goal_mask = in_dynamic_horizon & valid_actor_team & (df_events['next_goal_team_id'] != df_events['team_id'])

df_events.loc[poss_goal_mask, 'target'] = CLASS_POSS_GOAL
df_events.loc[def_goal_mask, 'target'] = CLASS_DEF_GOAL

# ------------------------------------------------------------------
# 4) Diagnostics / invariants
# ------------------------------------------------------------------
print("\nHorizon diagnostics:")
print(f"  Rows with horizon_end_time: {df_events['horizon_end_time'].notna().sum():,} / {len(df_events):,}")
print(f"  Penalty-window events (penalty + penaltydrawn): {is_penalty_window_event.sum():,}")
print(f"  Penalty-window rows with +120s override: {has_faceoff_for_penalty.sum():,}")
print(f"  Penalty-window rows missing next faceoff (kept default +20s): {(is_penalty_window_event & df_events['next_faceoff_time'].isna()).sum():,}")
print(f"  Rows with sec_next_stoppage: {df_events['sec_next_stoppage'].notna().sum():,}")

# Ensure no cross-period horizon overflow
overflow = int((df_events['horizon_end_time'] > df_events['period_cap_time']).sum())
print(f"  Horizon overflow rows (should be 0): {overflow}")

# Ensure earliest-goal priority creates mutually exclusive class assignment
both_goal_classes = int((poss_goal_mask & def_goal_mask).sum())
print(f"  Rows marked both poss+def (should be 0): {both_goal_classes}")

# Additional invariants
negative_stoppage = int((df_events['sec_next_stoppage'] < 0).fillna(False).sum())
print(f"  Negative sec_next_stoppage rows (should be 0): {negative_stoppage}")

invalid_targets = int((~df_events['target'].isin([CLASS_POSS_GOAL, CLASS_DEF_GOAL, CLASS_NO_GOAL])).sum())
print(f"  Invalid target labels (should be 0): {invalid_targets}")

print(f"\n{'='*80}")
print("TARGET DISTRIBUTION (DYNAMIC HORIZON, ACTOR-RELATIVE)")
print(f"{'='*80}")

target_counts = df_events['target'].value_counts().sort_index()
target_pcts = df_events['target'].value_counts(normalize=True).sort_index() * 100

target_names = {
    CLASS_POSS_GOAL: 'Possession Goal',
    CLASS_DEF_GOAL: 'Defending Goal',
    CLASS_NO_GOAL: 'No Goal',
 }

for target_class in sorted(target_names.keys()):
    class_name = target_names[target_class]
    count = int(target_counts.get(target_class, 0))
    pct = float(target_pcts.get(target_class, 0.0))
    print(f"  Class {target_class} ({class_name:15s}): {count:>9,} ({pct:>5.2f}%)")

print(f"\n✓ Added dynamic horizon columns: horizon_end_time, sec_next_stoppage")
print("✓ Applied earliest-goal priority within dynamic window (no dual-class conflicts)")
print("=" * 80)

# cleanup helper
if '_row_id' in df_events.columns:
    df_events.drop(columns=['_row_id'], inplace=True)
```

### Goal Attribution Logic: Creating `target_xg` for xG Model

**Objective**: Create a binary label for the Transformer xG model that identifies shots/deflections that directly resulted in goals.

**Logic**:
1. Implement vectorized look-ahead to attribute goals to the **last shot or deflection** before the goal
2. **Filter out noise**: Ignore failed block events when determining scoring events
3. Rule: If `event_t+1` is a goal (or `event_t+2` if `t+1` is a failed block), and `event_t` is a shot/deflection, then `target_xg = 1`

**Why This Matters**:
- Traditional xG looks at a snapshot: Distance, Angle, Shot Type
- Our Transformer xG will capture **pre-shot context**: Fatigue, passing velocity, defensive pressure over the last 128 events
- This binary target allows direct comparison with `sl_xg_all_shots` as a baseline validation

```python

print(f"\nsl_xg_all_shots statistics:")
print(df_events['sl_xg_all_shots'].describe())

# Check for missing values
print(f"\nMissing sl_xg_all_shots values: {df_events['sl_xg_all_shots'].isna().sum():,} ({df_events['sl_xg_all_shots'].isna().mean()*100:.2f}%)")
print(f"  Events with sl_xg_all_shots: {df_events['sl_xg_all_shots'].notna().sum():,} ({df_events['sl_xg_all_shots'].notna().mean()*100:.2f}%)")
print(f"  sl_xg_all_shots range: [{df_events['sl_xg_all_shots'].min():.4f}, {df_events['sl_xg_all_shots'].max():.4f}]")
```

```python
# Vectorized look-ahead logic to create target_xg
print("=" * 80)
print("CREATING BINARY TARGET_XG FOR TRANSFORMER XG MODEL")
print("=" * 80)

# ------------------------------------------------------------------
# 1) Preserve current event order unless abnormal temporal drift exists
#    Abnormal = period_time delta < -1 within (game_id, period, sequence_id)
# ------------------------------------------------------------------
required_cols = ['game_id', 'period', 'sequence_id', 'period_time', 'event_type', 'outcome']
missing_cols = [c for c in required_cols if c not in df_events.columns]
if missing_cols:
    raise KeyError(f"Missing required columns for xG target creation: {missing_cols}")

df_events['period_time'] = pd.to_numeric(df_events['period_time'], errors='coerce')
seq_keys = ['game_id', 'period', 'sequence_id']

delta = df_events.groupby(seq_keys, sort=False)['period_time'].diff()
abnormal_mask = delta < -1
n_abnormal = int(abnormal_mask.sum())
n_abnormal_seq = int(df_events.loc[abnormal_mask, seq_keys].drop_duplicates().shape[0])

print(f"Abnormal period_time deltas (< -1): {n_abnormal:,} across {n_abnormal_seq:,} sequences")

if n_abnormal > 0:
    print("Applying strict goal-adjacency reorder only because abnormality was detected...")

    import gc

    # Stable baseline ordering inside sequence for repair pass
    df_events.sort_values(['game_id', 'period', 'sequence_id', 'period_time'], kind='mergesort', inplace=True)
    df_events.reset_index(drop=True, inplace=True)
    df_events['_sort_key'] = df_events.index.astype(float)

    # Movers: event directly before goal that should be moved away from goal adjacency
    type_mover = {'lpr', 'penalty', 'penaltydrawn'}
    # Anchors: shot chain events that should remain adjacent to goal
    type_anchor = {'shot', 'deflection', 'block'}

    goal_indices = df_events.index[df_events['event_type'] == 'goal'].to_numpy()
    count_moved = 0

    for goal_idx in goal_indices:
        prev_idx = goal_idx - 1
        if prev_idx < 0:
            continue

        # Must stay inside sequence
        if df_events.at[prev_idx, 'sequence_id'] != df_events.at[goal_idx, 'sequence_id']:
            continue

        # Only act when goal-1 is a mover
        if df_events.at[prev_idx, 'event_type'] not in type_mover:
            continue

        mover_idx = prev_idx
        anchor_idx = None

        # Search up to 5 events backward from mover for shot-chain anchor
        start_search = max(0, mover_idx - 5)
        for curr_idx in range(mover_idx - 1, start_search - 1, -1):
            if df_events.at[curr_idx, 'sequence_id'] != df_events.at[goal_idx, 'sequence_id']:
                break

            etype = df_events.at[curr_idx, 'event_type']
            if etype in type_anchor:
                anchor_idx = curr_idx  # keep earliest valid in this short backward scan
            elif etype in type_mover:
                pass
            else:
                break

        if anchor_idx is not None:
            new_key = float(anchor_idx) - 0.5
            target_time = df_events.at[anchor_idx, 'period_time']

            df_events.at[mover_idx, '_sort_key'] = new_key
            df_events.at[mover_idx, 'period_time'] = target_time
            count_moved += 1

    if count_moved > 0:
        print(f"✓ Moved & synced {count_moved:,} mover events")
        df_events = df_events.sort_values(['game_id', 'period', 'sequence_id', '_sort_key'], kind='mergesort').reset_index(drop=True)
    else:
        print("✓ No mover events required goal-adjacency correction")

    df_events.drop(columns=['_sort_key'], inplace=True)
    gc.collect()
else:
    print("No abnormality detected; preserving current event ordering (no resort).")
    df_events = df_events.reset_index(drop=True)

# ------------------------------------------------------------------
# 2) Build target_xg from immediate look-ahead in preserved/corrected order
# ------------------------------------------------------------------
df_events['target_xg'] = 0

# Identify shots and deflections (potential scoring events)
shot_events = df_events['event_type'].isin(['shot', 'deflection'])

# Identify failed blocks
failed_block = (df_events['event_type'] == 'block') & (df_events['outcome'] == 'failed')

# Create shifted columns for look-ahead
df_events['next_event_type'] = df_events['event_type'].shift(-1)
df_events['next_outcome'] = df_events['outcome'].shift(-1)
df_events['next_next_event_type'] = df_events['event_type'].shift(-2)

# Track sequence boundaries (to avoid looking across sequences)
df_events['next_sequence_id'] = df_events['sequence_id'].shift(-1)
df_events['next_next_sequence_id'] = df_events['sequence_id'].shift(-2)
df_events['same_sequence_next'] = df_events['sequence_id'] == df_events['next_sequence_id']
df_events['same_sequence_next_next'] = df_events['sequence_id'] == df_events['next_next_sequence_id']

# Case 1: Shot/deflection followed directly by goal (no noise)
direct_scoring = (
    shot_events
    & (df_events['next_event_type'] == 'goal')
    & df_events['same_sequence_next']
 )

# Case 2: Shot/deflection followed by failed block, then goal
blocked_scoring = (
    shot_events
    & (df_events['next_event_type'] == 'block')
    & (df_events['next_outcome'] == 'failed')
    & (df_events['next_next_event_type'] == 'goal')
    & df_events['same_sequence_next_next']
 )

# Assign target_xg = 1 for scoring shots
df_events.loc[direct_scoring | blocked_scoring, 'target_xg'] = 1

# Clean up helper columns
df_events.drop(columns=[
    'next_event_type', 'next_outcome', 'next_next_event_type',
    'next_sequence_id', 'next_next_sequence_id',
    'same_sequence_next', 'same_sequence_next_next'
], inplace=True)

# Statistics
print(f"\nSHOT/DEFLECTION EVENT BREAKDOWN:")
print(f"  Total shots/deflections: {shot_events.sum():,}")
print(f"  Scoring shots (target_xg=1): {df_events['target_xg'].sum():,}")
print(f"  Non-scoring shots (target_xg=0): {(shot_events & (df_events['target_xg'] == 0)).sum():,}")
print(f"  Scoring rate: {df_events['target_xg'].sum() / max(1, shot_events.sum()) * 100:.2f}%")

print(f"\nDIRECT VS. BLOCKED SCORING SHOTS:")
print(f"  Direct scoring (shot → goal): {direct_scoring.sum():,}")
print(f"  Blocked scoring (shot → failed block → goal): {blocked_scoring.sum():,}")

print(f"\nTARGET_XG DISTRIBUTION:")
print(df_events['target_xg'].value_counts().sort_index())

print(f"\n✓ Created target_xg column for xG model training")
print("=" * 80)
```

### Create Counterfactual Outcome Column for xG Model

**Rationale**: The xG model is counterfactual — it predicts shot quality based on positioning, defensive pressure, and pre-shot context, **regardless of whether the shot was on-net or missed wide**.

For shots and deflections, we mask (set to NaN) the outcome column so the model can't "cheat" by using whether the shot was on-net.

For all other events, we preserve the outcome to provide additional context (pass was completed/incomplete, block was successful/failed, etc.).

**Result**: 
- Create `outcome_xg` column with outcome masked for shots/deflections
- Keep original `outcome` column intact for Optimus Reim (Phase 5)

### Create Actor-Relative Targets (Dynamic Horizon, Vectorized)

Three-class classification where each event is labeled relative to its own team_id using a dynamic, stoppage-aware horizon:
- Class 0 (Possession Goal): Earliest future goal in horizon is by event team
- Class 1 (Defending Goal): Earliest future goal in horizon is by opposing team
- Class 2 (No Goal): No goal in dynamic horizon

Dynamic horizon rules:
- Standard events: period_time + 20s
- Penalties: next faceoff + 120s (single default window for all penalties)
- Capped at period boundary to prevent intermission bleed
- sec_next_stoppage stores seconds to next whistle/goal/end_of_period

Earliest-goal priority guarantees mutual exclusivity (no event can be both Class 0 and Class 1).

### Build Vocabularies for Categorical Features

Create integer encodings for event_type, outcome, and period. Reserve 0 for padding token, 1 for unknown.
